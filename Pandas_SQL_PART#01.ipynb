```python
import pandas as pd
import pandasql as pd_sql
```

```python
#for practice, I'll be using titanic dataset
#link to the dataset https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv
#reading data from dataset using pandas read_csv
titanic_ds = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
#to show some top rows from titanic dataset 
titanic_ds.head()
#to show some bottom rows from titanic dataset 
titanic_ds.tail()
```

```python
#checking what is available in whole data_set, we need to write an sql query, pandasql will help for that
pd_sql.sqldf('''
select * from titanic_ds
''')
#similarly if the table name is too much long then we can give an alias to it as following
pd_sql.sqldf('''
select t_ds.* from titanic_ds as t_ds
'''
)
```

```python
#Usually datasets are very large that's why we sometimes want to get specific amount of rows we can do it as
pd_sql.sqldf('''
select t_ds.* from titanic_ds as t_ds limit 10
'''
)
#only info of the top 10 records will be shown
```

```python
pd_sql.sqldf('''
select t_ds.* from titanic_ds as t_ds limit 1,5
''')
```

```python
#top 5 people with highest age
pd_sql.sqldf('''
select t_ds.name,t_ds.age from titanic_ds as t_ds order by age desc limit 5
''')
#find record of a person with 3rd highest age
pd_sql.sqldf('''
select t_ds.name,t_ds.age from titanic_ds as t_ds order by age desc limit 2,1
'''
)
```

```python
#top 50 record of the passengers with gender male and below age 40 or 
#gender female and age greater than 20 less than 40 order in decreasing by age
pd_sql.sqldf('''
select t_ds.passengerId,t_ds.name,t_ds.sex, t_ds.age, t_ds.ticket
from titanic_ds as t_ds
where (t_ds.age < 40 and t_ds.sex = 'male')
or ((t_ds.age > 20 and t_ds.age < 40) and t_ds.sex = 'female')
order by t_ds.age desc
limit 50
''')
```

```python
#show the records of passengers with age between 15 to 40 and class of passenger 1 or 3
pd_sql.sqldf('''
select t_ds.passengerId,t_ds.pclass,t_ds.name,t_ds.sex, t_ds.age, t_ds.ticket
from titanic_ds as t_ds
where (t_ds.age between 15 and 40)
and
t_ds.pclass in (1,3)
''')
```

```python
#Let's say that we have no sex column in our data set then how will we be predicting the records of males and females
#we'll using concept of SQL like in this case on name column
#for male
pd_sql.sqldf('''
select t_ds.passengerId,t_ds.pclass,t_ds.name, t_ds.age, t_ds.ticket
from titanic_ds as t_ds
where t_ds.name like '%, Mr.%'
''')
#for female
pd_sql.sqldf('''
select t_ds.passengerId,t_ds.pclass,t_ds.name, t_ds.age, t_ds.ticket
from titanic_ds as t_ds
where t_ds.name like '%, Mrs.%'
''')
```

```python
#Now say we want to know the count of male and female who boarded the ship
pd_sql.sqldf('''
select count(t_ds.sex) total_passengers, t_ds.sex Gender
from titanic_ds as t_ds
group by t_ds.sex
order by t_ds.sex desc
''')
```

```python
#get the record of passenger with max and min age
#for highest age
pd_sql.sqldf('''
select t_ds.name,max(t_ds.age) highest_age
from titanic_ds as t_ds
''')
#for lowest age
pd_sql.sqldf('''
select t_ds.name,min(t_ds.age) lowest_age
from titanic_ds as t_ds
''')
```

```python
#Divide the age into 6 categories
#1. 55+ old 2. < 55 and > 38 adults 3. <= 38 and > 22 young 4. <=22 and >= 15 teenager  5. less than 15 children 6. else invalid 

#we'll do this by using case when concept of SQL
pd_sql.sqldf('''
select t_ds.passengerId,t_ds.pclass,t_ds.name, t_ds.age, 
case
    when (t_ds.age > 55 and t_ds.age <= 130) then 'OLD'
    when (t_ds.age <= 55 and t_ds.age > 38) then 'ADULT'
    when (t_ds.age <= 38 and t_ds.age > 22) then 'YOUNG'
    when (t_ds.age <= 22 and t_ds.age >= 15) then 'TEENAGER'
    when t_ds.age <15 then 'CHILD'
    else 'INVALID'
end as 'AGE_GROUP'
from titanic_ds as t_ds
'''
)
```

```python
#Now lets say that you need to extract some pattern from a specific column
#we'll be using substr fn of SQL substr(1,5) from 1st letter till 5th including 5th
pd_sql.sqldf('''
select t_ds.name, substr(t_ds.name,1,5) as extracted_name
from titanic_ds as t_ds
''')
```

```python
#now let us say, we want to show names of passengers in lowercase and gender in uppercase
#we'll be using upper() and lower() fns
pd_sql.sqldf('''
select lower(t_ds.name) Name, upper(t_ds.sex) Gender
from titanic_ds as t_ds
''')
```

```python
#let's say we are interested in getting info about average_age of male/female in our dataset
pd_sql.sqldf('''
select count(t_ds.sex) Total_Passengers,avg(t_ds.age) Average_Age, t_ds.sex Gender
from titanic_ds as t_ds
group by t_ds.sex
order by t_ds.sex desc
''')
```
